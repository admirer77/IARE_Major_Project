```python

df = df.rename(columns={' Label': 'Label'})

print(df.columns)
```

```python
unique_labels = df['Label'].unique()
print("\nUnique values in 'Label' column:")
print(unique_labels)
```

```python
# prompt: create new column Label_int and map 'NetBIOS' 'BENIGN' 'LDAP' 'MSSQL' 'Portmap' 'Syn' 'UDP' 'UDPLag' in Label with numbers

label_mapping = {
    'NetBIOS': 0,
    'BENIGN': 1,
    'LDAP': 2,
    'MSSQL': 3,
    'Portmap': 4,
    'Syn': 5,
    'UDP': 6,
    'UDPLag': 7
}

df['Label_int'] = df['Label'].map(label_mapping)

df
```

```python


categorical_cols = df.select_dtypes(include=['object']).columns

print("\nCategorical columns:")
print(categorical_cols)
```

```python
numerical_cols = df.select_dtypes(include=['number']).columns

print("\nNumerical columns:")
print(numerical_cols)
```

```python
# prompt: create another dataframe with only numerical columns

numerical_df = df[numerical_cols]
numerical_df
```

```python
# prompt: rows and cols in df

num_rows_ = numerical_df.shape[0]
num_cols_ = numerical_df.shape[1]

print("Number of rows:", num_rows_)
print("Number of columns:", num_cols_)
```
